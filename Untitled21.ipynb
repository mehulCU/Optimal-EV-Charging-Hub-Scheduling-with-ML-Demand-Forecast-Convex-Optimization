```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load data (ensure this path matches Colab file upload)
df = pd.read_csv('ev_charging_bangalore_complete.csv', parse_dates=['timestamp'])

# Example: filter to two hubs (0 and 1) for demonstration
df = df[df.hub_id.isin([0,1])]

# Sort by time for each hub
df = df.sort_values(['hub_id','timestamp'])

# Extract temporal features
df['hour'] = df['timestamp'].dt.hour
df['dayofweek'] = df['timestamp'].dt.dayofweek
df['is_weekend'] = (df['dayofweek'] >= 5).astype(int)
df['month'] = df['timestamp'].dt.month

# The dataset already has 'is_peak_hour', 'is_rainy'; we keep them
# Ensure we have one-hot or numeric encoding where needed
df['is_peak_hour'] = df['is_peak_hour'].astype(int)
df['is_rainy'] = df['is_rainy'].astype(int)

# Lag and rolling features: use demand_vehicles from previous interval(s)
df['demand_prev_1'] = df.groupby('hub_id')['demand_vehicles'].shift(1).fillna(0)
df['demand_prev_24'] = df.groupby('hub_id')['demand_vehicles'].shift(24).fillna(0)  # 6h lag if 15-min intervals

# Rolling mean demand for last 4 hours (if 15-min intervals, that's 16 points)
df['demand_roll_4h'] = df.groupby('hub_id')['demand_vehicles'].transform(lambda x: x.shift(1).rolling(16).mean().fillna(0))

# Select features for modeling
features = ['hour','dayofweek','is_weekend','month',
            'is_peak_hour','is_rainy',
            'temperature_celsius','humidity_percent',
            'demand_prev_1','demand_prev_24','demand_roll_4h']
X = df[features]
y = df['demand_vehicles']

# Split into train/test sets (e.g., last 7 days for testing)
train_df, test_df = train_test_split(df, test_size=0.1, shuffle=False)
X_train = train_df[features]
y_train = train_df['demand_vehicles']
X_test = test_df[features]
y_test = test_df['demand_vehicles']

# Normalize continuous features
scaler = StandardScaler()
cont_features = ['temperature_celsius','humidity_percent','demand_prev_1','demand_prev_24','demand_roll_4h']
X_train[cont_features] = scaler.fit_transform(X_train[cont_features])
X_test[cont_features] = scaler.transform(X_test[cont_features])
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Train XGBoost Regressor
model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    random_state=42
)
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate performance
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"XGBoost Forecast Results:")
print(f"R² Score: {r2:.3f}")
print(f"RMSE: {rmse:.3f} vehicles")
```

```python
import cvxpy as cp

# --- Problem setup ---
P_charger = 11.0  # kW per charger
m = np.array([20, 15])  # number of chargers at site 0 and site 1
P_site_max = m * P_charger
P_grid_max = 100.0  # total grid capacity in kW

# Time horizon
T = demand_site.shape[0]
S = demand_site.shape[1]

# Decision variables
p = cp.Variable((T, S))  # power delivered [kW]
w = cp.Variable((T, S))  # waiting vehicles

constraints = []

# Initial queue (assume zero at t=0)
constraints.append(w[0, :] == demand_site[0, :] - p[0, :] / P_charger)

# Queue dynamics and power constraints
for t in range(1, T):
    constraints.append(
        w[t, :] == w[t - 1, :] + demand_site[t, :] - p[t, :] / P_charger
    )

constraints += [
    p >= 0,
    w >= 0,
    p <= P_site_max,  # per site max
    cp.sum(p, axis=1) <= P_grid_max  # total grid constraint
]

# --- Objective function: load variance + wait time ---
P_total = cp.sum(p, axis=1)
P_avg = cp.sum(P_total) / T
V = cp.sum_squares(P_total - P_avg) / T  # variance of load
W = cp.sum(w) / T  # average wait (vehicles)

# Weights for trade-off
lambda_var = 1.0
lambda_wait = 1.0

objective = cp.Minimize(lambda_var * V + lambda_wait * W)
problem = cp.Problem(objective, constraints)

# --- Solve ---
result = problem.solve(solver=cp.SCS, verbose=False)

# --- Extract solution ---
p_opt = p.value
w_opt = w.value

print(f"Optimization completed.")
print(f"Peak Load: {np.max(np.sum(p_opt, axis=1)):.2f} kW")
print(f"Total Wait: {np.sum(w_opt):.1f} vehicles across {S} sites and {T} intervals.")
```

```python
# Prepare arrays
p_fcfs = np.zeros_like(p_opt)
p_heur = np.zeros_like(p_opt)
w_fcfs = np.zeros_like(w_opt)
w_heur = np.zeros_like(w_opt)
queue_fcfs = np.zeros(S)
queue_heur = np.zeros(S)

for t in range(T):
    arrivals = demand_site[t].copy()
    # FCFS: serve up to m_s each site
    for s in range(S):
        queue_fcfs[s] += arrivals[s]
        served = min(queue_fcfs[s], m[s])
        p_fcfs[t,s] = served * P_charger
        queue_fcfs[s] -= served
        w_fcfs[t,s] = queue_fcfs[s]
    # Heuristic: allocate power proportionally
    total_arr = sum(arrivals)
    for s in range(S):
        queue_heur[s] = arrivals[s]  # start new queue (no carry-over)
    if total_arr > 0:
        for s in range(S):
            alloc = (arrivals[s]/total_arr)*P_grid_max
            alloc = min(alloc, P_site_max[s])
            served = min(arrivals[s], alloc / P_charger)
            p_heur[t,s] = served * P_charger
            queue_heur[s] = arrivals[s] - served
            w_heur[t,s] = queue_heur[s]

# Compute metrics
total_load_fcfs = np.sum(p_fcfs, axis=1)
total_load_heur = np.sum(p_heur, axis=1)
V_fcfs = np.var(total_load_fcfs)
V_heur = np.var(total_load_heur)
W_fcfs = np.sum(w_fcfs)
W_heur = np.sum(w_heur)
print(f"FCFS: Load Variance={V_fcfs:.1f}, Total Wait={W_fcfs:.1f}")
print(f"Heuristic: Load Variance={V_heur:.1f}, Total Wait={W_heur:.1f}")
```
